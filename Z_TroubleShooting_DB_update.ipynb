```python
#check DBTP creation base on stock list's create log
import subprocess
import linecache
import io
import re

def get_last_stock_date(processID):
    fnwp='/home/rdchujf/n_workspace/data/RL_data/I_DB/Stock_List/SLV300/CreateLog/Process{0}Output.txt'.format(processID)
    cmd = ['awk','/start process at/{ print NR }' ,fnwp]
    try:
        linenumbers = subprocess.run(cmd, stdout=subprocess.PIPE)
        f=io.BytesIO(linenumbers.stdout)
        f.readline()
        Number_of_Line=int(f.readline().decode("utf-8") )
    except:

        line = subprocess.check_output(['tail', '-1', fnwp])
        if line.decode("utf-8")=="End with Success\n":
            return True
        else:
            print (fnwp, line)
            return False
    line=linecache.getline(fnwp, Number_of_Line-2)
    if line=="End with Success\n":
        return True
    else:
        results=re.findall(r'DBTP from DBI (\w+)\W(\w+)',line)
        stock, dateS=results[0]
        print (fnwp)
        print (stock, dateS)
        return False

for processID in list(range(0,32)):
    get_last_stock_date(processID)
```
