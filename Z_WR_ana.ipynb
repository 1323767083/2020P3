```python
import os
import pandas as pd
system_name="again_try_resid_lr001_scal20"
process_name="EvalAgent_4"
dnwp=os.path.join("/mnt/data_disk2/n_workspace/RL/",system_name,"WR",process_name)

ETs=[int(fn[2:-4]) for fn in os.listdir(dnwp)]
ETs.sort()

for ET in ETs:
    fwnp=os.path.join(dnwp,"ET{0}.csv".format(ET))
    df=pd.read_csv(fwnp)
    df["buy"]=df["BW"]+df["BZ"]+df["BR"]
    df["all"]=df["BW"]+df["BZ"]+df["BR"]+df["NW"]+df["NZ"]+df["NR"]+df["NA"]
    assert len(set(df["all"].to_list()))==1, df["all"]
    print (ET,df["BW"].sum()/df["buy"].sum(),df["BR"].sum()/df["buy"].sum(),(df["BR"].sum()+df["BZ"].sum())/df["buy"].sum(),df["buy"].sum()/ df["all"].sum())
```
